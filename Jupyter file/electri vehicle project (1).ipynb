```python
#Which electric utilities serve the largest number of EVs?
#Visulization:
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Electric_Vehicle_Population_Data.csv")
plt.figure(figsize=(10,6))
sns.countplot(
    y="Electric Utility",
    data=df,
    order=df["Electric Utility"].value_counts().head(10).index,
    hue="Electric Utility", 
    palette="viridis",
    legend=False)
plt.title("Top 10 Electric Utilities Serving EVs")
plt.xlabel("Number of EVs")
plt.ylabel("Electric Utility")
plt.show()
```

```python
#What is the growth trend of electric vehicles by year?
#visualization:
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Electric_Vehicle_Population_Data.csv")
plt.figure(figsize=(10,6))
sns.lineplot(
    x=df["Model Year"].value_counts().sort_index().index,
    y=df["Model Year"].value_counts().sort_index().values,
    marker="o", color="red"
)
plt.title("Growth Trend of Electric Vehicles by Year")
plt.xlabel("Model Year")
plt.ylabel("Number of EVs")
plt.grid(True)
plt.show()
```

```python
#How has the number of EVs per state changed over time?
#visualization:
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Electric_Vehicle_Population_Data.csv")
plt.figure(figsize=(12,6))
sns.lineplot(
    x="Model Year", y="Count", hue="State",
    data=df.groupby(["State","Model Year"]).size().reset_index(name="Count")
)
plt.title("Number of EVs per State Over Time")
plt.xlabel("Model Year")
plt.ylabel("Number of EVs")
plt.legend(title="State", bbox_to_anchor=(1.05,1), loc="upper left")
plt.grid(True)
plt.show()
```

```python
#Do vehicles with longer electric ranges tend to be concentrated in certain states or cities?
#visualization:states
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Electric_Vehicle_Population_Data.csv")
plt.figure(figsize=(10,6))
sns.barplot(
    x="Electric Range", 
    y="State", 
    data=df, 
    estimator="mean", 
    hue="State",           
    palette="viridis", 
    legend=False,             
    order=df.groupby("State")["Electric Range"].mean().sort_values(ascending=False).head(10).index
)
plt.title("Top 10 States by Average EV Range")
plt.xlabel("Average Electric Range (miles)")
plt.ylabel("State")
plt.show()
```
